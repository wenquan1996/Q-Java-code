```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import warnings
import random
from datetime import datetime
random.seed(datetime.now())
warnings.filterwarnings('ignore')
import sqlite3
```

```python
Domain=pd.read_csv('Players_Details.csv')
Company=pd.read_csv('NBA_Team_Accout_Twitter.csv')
Producer=pd.read_csv('Player_Producer.csv')
Consumer=pd.read_csv('HashTag_Players.csv')
DomainToConsumer=pd.read_csv('Domain_Consumer.csv')
DomainToProducer=pd.read_csv('Domain_Producer.csv')
NBA_db='NBA_Database.sqlite'
conn=sqlite3.connect(NBA_db)
c=conn.cursor()

Domain.to_sql('Domain',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False) 
Company.to_sql('Company',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False)
Producer.to_sql('Producer',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False)
Consumer.to_sql('Consumer',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False)
DomainToConsumer.to_sql('DomainToConsumer',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False)
DomainToProducer.to_sql('DomainToProducer',             # Name of the table.
             con=conn,                    # The handle to the file that is set up.
             if_exists='replace',         # Overwrite, append, or fail.
             index=False)
```
