```python
#import library
import instaloader
import pandas as pd
import csv
```

```python

name=[]
socialid=[]
likes=[]
comments=[]
caption=[]
hashtag=[]
mediaid=[]
i=0
n="Kevin"
L=instaloader.Instaloader()
#get user profile according to  player's name and from post to get attributes
posts = L.get_hashtag_posts("kevindurantedit")
users=set()
for post in posts:
    if i <500:
        users.add(post.owner_profile)
        table_1=post.owner_username
        name.append(table_1)
        table_2=post.owner_id
        socialid.append(table_2)
        table_3=post.likes
        likes.append(table_3)
        table_4=post.comments
        comments.append(table_4)
        table_5=post.caption
        caption.append(table_5)
        table_6=post.caption_hashtags
        hashtag.append(table_6)
        table_7=post.mediaid
        mediaid.append(table_7)
        i+=1
```

```python
#create dataframe
user_info={'name':name,'id':socialid,'likes':likes,'comments':comments,'caption':caption,'hashtag':hashtag,'mediaid':mediaid}
df=pd.DataFrame(user_info)
df
```

```python
#export csv file and extract hashtags
df.to_csv("kevin_topic.csv")
df1=pd.read_csv("kevin_topic.csv")
df2=df1.drop('hashtag',axis=1).join(df1["hashtag"].str.split(',',expand=True).stack().reset_index(level=1,drop = True).rename("tags"))
df2.to_csv("kevin_topic_ins.csv")
df2
```
