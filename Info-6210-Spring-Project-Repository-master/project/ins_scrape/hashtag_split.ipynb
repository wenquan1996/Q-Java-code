```python
import pandas as pd
import csv
```

```python
df=pd.read_csv("social_hashtag_except_split.csv")
```

```python
#split hashtags according to ","
df1=df.drop('hashtag',axis=1).join(df["hashtag"].str.split(',',expand=True).stack().reset_index(level=1,drop = True).rename("tags"))
```

```python
df1.to_csv("social_hashtags_with_null.csv")
```
